```python
seoul = pd.read_csv('기타휴게음식점_정제.csv')

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

# 서울시 전체 대상으로 '영업상태명'이 '영업/정상'인 카페의 전체 연도별 분포 시각화
open_business = seoul[(seoul['영업상태명'] == '폐업') & (seoul['폐업_년'])]

plt.figure(figsize=(8, 7))
sns.countplot(y=open_business['폐업_년'])
plt.xlabel('Count')
plt.ylabel('연도')
plt.title('서울시 폐업 카페 연도별 분포')
plt.show()
```

```python
# 서울시 전체 대상으로 '영업상태명'이 '영업/정상'인 카페의 전체 연도별 분포 시각화
open_business = seoul[(seoul['영업상태명'] == '영업/정상') & (seoul['인허가_년'])]

plt.figure(figsize=(8, 6))
sns.countplot(y=open_business['인허가_년'])
plt.xlabel('Count')
plt.ylabel('연도')
plt.title('서울시 인허가 카페 연도별 분포')
plt.show()
```

```python
# 서울시 전체 대상으로 '영업상태명'이 '폐업'이고, '영업년도'가 2021년부터 2023년까지인 행들을 필터링
closed_business = seoul[(seoul['영업상태명'] == '폐업') & (seoul['폐업_년'].between(2021, 2023))]

# 2021년부터 2023년까지의 서울시 인허 카페 연도별 분포 시각화
plt.figure(figsize=(8, 3))
sns.countplot(y=closed_business['폐업_년'], order=closed_business['폐업_년'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('연도')
plt.title('서울시 폐업 카페 연도별 분포')
plt.show();
```

```python
# 서울시 전체 대상으로 '영업상태명'이 '영업/정상'이고, '인허가 연도'가 2021년부터 2023년까지인 행들을 필터링
open_business = seoul[(seoul['영업상태명'] == '영업/정상') & (seoul['인허가_년'].between(2021, 2023))]

# 2021년부터 2023년까지의 서울시 인허가 카페 연도별 분포 시각화
plt.figure(figsize=(8, 3))
sns.countplot(y=open_business['인허가_년'], order=open_business['인허가_년'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('연도')
plt.title('서울시 인허가 카페 연도별 분포')
plt.show()
```

```python
# '영업상태명'이 '폐업'이고, '영업년도'가 2021년부터 2023년까지인 행들을 필터링
closed_business_21_23 = seoul[(seoul['영업상태명'] == '폐업') & (seoul['폐업_년'].between(2021, 2023))]

# 행정구역별로 카운트하여 시각화
sns.countplot(y=closed_business_21_23['구'], order=closed_business_21_23['구'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('구')
plt.title('서울시 2021년부터 2023년까지 폐업 카페 행정구역별 분포')
plt.show()
```

```python
# '영업상태명'이 '영업/정상'이고, '인허가 연도'가 2021년부터 2023년까지인 행들을 필터링
open_business_21_23 = seoul[(seoul['영업상태명'] == '영업/정상') & (seoul['인허가_년'].between(2021, 2023))]

# 행정구역별로 카운트하여 시각화
sns.countplot(y=open_business_21_23['구'], order=open_business_21_23['구'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('구')
plt.title('서울시 2021년부터 2023년까지 인허가 카페 행정구역별 분포')
plt.show()
```

```python
# '영업상태명'이 '폐업'이고, '폐업 연도'가 2022년인 행들을 필터링
closed_business_22 = seoul[(seoul['영업상태명'] == '폐업') & (seoul['폐업_년'] == 2022)]

# 행정구역별로 카운트하여 시각화
sns.countplot(y=closed_business_22['구'], order=closed_business_22['구'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('구')
plt.title('서울시 2022년 폐업 카페 행정구역별 분포')
plt.show()
```

```python
# '영업상태명'이 '영업/정상'이고, '인허가 연도'가 2022년인 행들을 필터링
open_business_22 = seoul[(seoul['영업상태명'] == '영업/정상') & (seoul['인허가_년'] == 2022)]

# 행정구역별로 카운트하여 시각화합니다.
sns.countplot(y=open_business_22['구'], order=open_business_22['구'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('구')
plt.title('서울시 2022년 인허가된 카페 행정구역별 분포')
plt.show()
```
